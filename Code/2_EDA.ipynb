## Problem understanding and definition
**Intelligent Diamond Reseller (IDR)**, a new diamond reselling business wants to innovate in the business, so they will use predictive modeling to estimate how much the market will pay for diamonds. Of course, to sell diamonds in the market, first they have to buy them from the producers; this is where predictive modeling becomes useful. Our work as data analysts/scientists is therefore to use the features contained in the dataset (except price which is our target) to:
1. Build a predictive model that predicts the price of diamonds, as accurately as possible &
2. Predict prices of diamonds offered to IDR by the producers, so IDR can decide how much to pay for those diamonds.

### Data Collection & Preparation
For this project, we don't need to get the data ourself using the extract, transform, load (ETL) technologies since IDL has been able to get a dataset (this is actually real-world data) containing the prices and key characteristics of about 54,000 diamonds. Find below the metadata about the dataset:
* price: Price in US dollars
* carat: Weight of the diamond
* cut: Quality of the cut (fair, good, very good, premium, ideal)
* color: Diamond color, from J (worst) to D (best)
* clarity: A measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
* x: Length in mm
* y: Width in mm
* z: Depth in mm
* depth: Total depth percentage = z / mean(x, y) = 2 * z / (x + y)
* table: Width of the top of the diamond relative to the widest point

With **Price(s)**, a continuous variable being the target we want to predict, it is obvious we are trying to solve a regression task, and will therefore build multiple regression models and select the best after evaluation for deployment.

### For the Numerical Features

```python
# We are going to use a histogram to visualize the distribution of the Numerical Features

num_features = ['price', 'carat', 'depth', 'table', 'x', 'y', 'z']
def desc_num_feature(feature, bins=30, edgecolor='k', **kwargs):
    fig, ax = plt.subplots(figsize=(8,4))
    df[feature].hist(bins=bins, edgecolor=edgecolor, ax=ax, **kwargs)
    ax.set_title(feature, size=15)
    desc = df[feature].describe().round(2).to_string()
    plt.figtext(1.0,0.15, desc, fontsize=17)
```

```python
for x in num_features:
    desc_num_feature(x)
```

#### Some observations from above:
* As **prices** increase, we observe fewer diamonds in our sample.
* A quarter(25%) of the diamond prices in our sample (around 13,500 - highest observation) are in that small interval of 326 - 949 USD, half (50%) the prices are below 2400 USD, three quarters (75%) of the prices are below 5300 USD, with the top quarter above that price.
* That is, excluding the bottom 25% and top 25%, the middle half of prices, i.e. between 950 - 5300 USD is the range where you can avoid purchasing either very cheap or very expensive diamonds.
* We can therefore also conclude that there isn't really a typical diamond price due to the large variability in prices.
* The **carat column**, has a max value 5.01 which is more than eigh standard deviations((5.01-1.04)/0.47)=8.45) from the 75th percentile, which is close to 1.04, & is a candidate for consideration as an outlier.
* For the 3 columns, **x (Length), y (Width), z (Depth)** representing the dimensions of the diamonds, the min values are zero, which is not possible since we would then be talking about a 1-D or 2-D diamonds, and will be considered as missing values. 
* Also the max of columns **y & z** are extremely high & can be considered outliers since it would be very hard to find a (normal) diamond of more than 3 centimeters (30 mm) in size in any of the dimensions, so if you observe such values, we can be sure that those are errors in measurement.

### For the Categorical Features

```python
cat_features = ['cut', 'color', 'clarity']
def plot_categorical(feature):
    count = df[feature].value_counts()
    percent = 100*df[feature].value_counts(normalize=True)
    dfc = pd.DataFrame({'count':count, 'percent':percent.round(1)})
    print(dfc)
    count.plot(kind='bar', title=feature + str(' feature'));
```

```python
plot_categorical('cut')
```

```python
plot_categorical('color')
```

```python
plot_categorical('clarity')
```

### Relationship Between The Features

```python
df[num_features].corr()
```

```python
# Lets look at the relationship between the numerical features
sns.pairplot(df[num_features], plot_kws={"s": 2}, diag_kind='kde');
```

```python
df.plot.scatter(x='depth', y='price', s=0.6)#, xlim=(50,75));
```

```python
# For numerical features & target
target = ['price']
for a in num_features:
    for b in target:
        _ = sns.scatterplot(x=a,y=b,data=df)
        plt.title('Relationship between ' +a+ ' and ' +b)
#       plt.savefig('../Images/Scatterplot of '+str(b)+' w.r.t '+str(a)+'.png')
        plt.show()
```

Some observations from above:
* We observe a nonlinear, though a clear pattern with definitely a lot of noise and a positive relationship between price and the following features, carat, x,y & z.
* x, y & z are stongly positively correlated to each other & we will apply a reduction technique before modeling since they essentially contain almost similar information.
* No clear pattern/relationship between depth & price, as well as table & price.

```python
# For categorical features & target
for a in cat_features:
    for b in target:
        _ = sns.boxplot(x=a,y=b,data=df)
        plt.title('Relationship between ' +a+ ' and ' +b)
        med = df.groupby(a)[b].agg(np.median).sort_values().to_string() #compare medians for the different categories
        plt.figtext(1.0,0.15, med, fontsize=17)
        #plt.savefig('../Images/Boxplot of '+str(b)+' w.r.t '+str(a)+'.png')
        plt.show()
```

```python
#lets use the order of the median prices to plot a more informative boxplot with rice < 10000

for a in cat_features:
    for b in target:
        med = df.groupby(a)[b].agg(np.median).sort_values() #compare medians for the different categories
        _ = sns.boxplot(x=a, y=b, data=df.loc[df['price']<10000], order=med.index)
        plt.title('Relationship between ' +a+ ' and ' +b)
        plt.figtext(1.0,0.15, med.to_string(), fontsize=17)
        #plt.savefig('../Images/Boxplot of '+str(b)+' w.r.t '+str(a)+'.png')
        plt.show()
```

Observations:
* The boxplots identify many outliers in all categories and features, which is essentially due to the distribution of prices.
* There is definitely a relationship between different cuts, color, clarity & the distribution of prices.
